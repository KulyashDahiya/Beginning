```python
def recursive_REVERSE(head):
    # Base case: if the list is empty or has only one node
    if head is None or head.next is None:
        return head
    
    # Recursive case: reverse the rest of the list
    small_head = recursive_REVERSE(head.next)
    
    # Reconnect the current node
    head.next.next = head  # Make the next node point to the current node
    head.next = None       # Disconnect the current node from the next node
    
    return small_head
```

```python
# Midpoint of a Linked List

def midPointLL(head):
    if not head:
        return None
    
    slow = head
    fast = head.next

    while fast and fast.next:
        # prev = slow
        slow = slow.next
        fast = fast.next.next

    return slow

head = takeInput_optimized("1 2 3 4 5 6 7")
printLL(head)
mid = midPointLL(head)
print(mid.data)
```

```python
## Merge 2 sorted LL

def MergeTwoSortedLL_self(head1, head2):

    final_head = Node(None)
    curr = final_head


    while head1 and head2:
        if head1.data <= head2.data:
            curr.next = Node(head1.data)
            head1 = head1.next
            curr = curr.next
        else:
            curr.next = Node(head2.data)
            head2 = head2.next
            curr = curr.next

    if head1:
        curr.next = head1
    else:
        curr.next = head2
    

    return final_head.next



head1 = takeInput_optimized("0 1 1 3 5 6 7")
head2 = takeInput_optimized("2 2 3 4 7")
printLL(head1)
printLL(head2)

head = MergeTwoSortedLL(head1, head2)
printLL(head)
```

```python
## Merge 2 sorted LL

def MergeTwoSortedLL(head1, head2):

    final_head = Node(None)
    curr = final_head


    while head1 and head2:
        if head1.data <= head2.data:
            curr.next = head1
            head1 = head1.next
        else:
            curr.next = head2
            head2 = head2.next
    
        curr = curr.next

    if head1:
        curr.next = head1
    else:
        curr.next = head2
    

    return final_head.next



head1 = takeInput_optimized("0 1 1 3 5 6 7")
head2 = takeInput_optimized("2 2 3 4 7")
printLL(head1)
printLL(head2)

head = MergeTwoSortedLL(head1, head2)
printLL(head)
```

```python
## MERGE SORT - Algo

def merge_sort(head):
    if not head or not head.next:
        return head
    
    mid = midPointLL(head)
    next_to_middle = mid.next

    mid.next = None

    left_half = merge_sort(head)
    right_half = merge_sort(next_to_middle)

    sorted_list = MergeTwoSortedLL(left_half, right_half)

    return sorted_list


head = takeInput_optimized("32 5 62 3 421 3 1 98 0 23")
printLL(head)
head = merge_sort(head)
printLL(head)
```

## Even after Odd Linked List

For a given singly linked list of integers, arrange the nodes such that all the even number nodes are placed after the all odd number nodes. The relative order of the odd and even terms should remain unchanged.
Note :
1. return the new head to the list.
2. Don't create a new linked list.
3. Just change the data, instead rearrange the provided list.

```python
## Even after Odd Linked List

def rearrange_odd_even(head):
    if not head or not head.next:
        return head

    # Initialize pointers for odd and even lists
    odd_head = None
    odd_tail = None
    even_head = None
    even_tail = None

    current = head

    # Traverse the list and separate odd and even nodes
    while current:
        if current.data % 2 != 0:  # Odd node
            if odd_head is None:
                odd_head = current
                odd_tail = odd_head
            else:
                odd_tail.next = current
                odd_tail = odd_tail.next
        else:  # Even node
            if even_head is None:
                even_head = current
                even_tail = even_head
            else:
                even_tail.next = current
                even_tail = even_tail.next

        current = current.next

    # If there are no odd nodes, return even_head directly
    if not odd_head:
        return even_head
    
    # If there are no even nodes, return odd_head directly
    if not even_head:
        return odd_head

    # Link odd list to even list
    odd_tail.next = even_head
    even_tail.next = None  # Mark the end of the combined list

    # Return the head of the new list which is the odd_head
    return odd_head


head = takeInput_optimized("1 2 3 4 5 6 7 -1")

print("Original List:")
printLL(head)

new_head = rearrange_odd_even(head)

print("Rearranged List:")
printLL(new_head)
```
