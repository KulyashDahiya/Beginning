```python
from google.cloud import pubsub_v1
import time
import random

# GCP Project and Pub/Sub topic details
project_id = "dataplatform-438202"
topic_id = "iot-data"

publisher = pubsub_v1.PublisherClient()
topic_path = publisher.topic_path(project_id, topic_id)

# Function to generate random sensor data
def generate_random_data():
    device_id = random.randint(1, 5)  # Random DeviceID between 1 and 5
    temperature = round(random.uniform(20.0, 30.0), 2)  # Temperature between 20.0 and 30.0
    pressure = round(random.uniform(1.0, 1.5), 2)  # Pressure between 1.0 and 1.5
    humidity = random.randint(30, 60)  # Humidity between 30 and 60
    wind_speed = round(random.uniform(3.0, 8.0), 2)  # Wind speed between 3.0 and 8.0

    message = f"DeviceID: {device_id}, Temperature: {temperature}, Pressure: {pressure}, Humidity: {humidity}, WindSpeed: {wind_speed}"
    return message

# Function to publish a message to Pub/Sub
def publish_message():
    message = generate_random_data()
    future = publisher.publish(topic_path, message.encode("utf-8"))
    print(f"Published message: {message}, Message ID: {future.result()}")

# Main loop to send messages continuously
if __name__ == "__main__":
    while True:
        publish_message()
        time.sleep(5)  # Send a message every 5 seconds
```

```python
import time
import random
import json
from google.cloud import pubsub_v1
from google.oauth2 import service_account

# GCP Project and Pub/Sub topic details
project_id = "dataplatform-438202"
topic_id = "iot-data"

# Create a Pub/Sub publisher client with default credentials (no need for credentials file)
publisher = pubsub_v1.PublisherClient()
topic_path = publisher.topic_path(project_id, topic_id)

# Function to generate random IoT sensor data in JSON format
def generate_random_data():
    data = {
        "device_id": random.randint(1, 10),  # Random DeviceID between 1 and 10
        "temperature": round(random.uniform(20.0, 35.0), 2),  # Temperature between 20°C and 35°C
        "humidity": random.randint(30, 90),  # Humidity between 30% and 90%
        "pressure": round(random.uniform(950, 1050), 2),  # Pressure between 950 and 1050 hPa
        "wind_speed": round(random.uniform(0.0, 20.0), 2),  # Wind speed between 0 and 20 m/s
        "timestamp": int(time.time())  # Current time as a timestamp
    }
    return data

# Function to publish the message to Pub/Sub
def publish_message():
    data = generate_random_data()  # Generate random sensor data
    # Convert the dictionary to a JSON string
    message_json = json.dumps(data)
    # Encode the JSON string and publish it to the topic
    future = publisher.publish(topic_path, message_json.encode("utf-8"))
    print(f"Published message: {message_json}, Message ID: {future.result()}")

# Main loop to continuously send data
if __name__ == "__main__":
    while True:
        publish_message()
        time.sleep(2)  # Simulate a delay of 2 seconds between messages
```
